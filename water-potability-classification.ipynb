```python
x=df.iloc[ : , :-1]
x.head()
```

```python
y=df.iloc[ : ,-1]
y.head()
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
print(f"Xtrain: {xtrain.shape}")
print(f"Xtest: {xtest.shape}")
print(f"Ytrain: {ytrain.shape}")
print(f"Ytest: {ytest.shape}")
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
xtrain=ss.fit_transform(xtrain)
xtest=ss.fit_transform(xtest)
```

```python
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
knn.fit(xtrain,ytrain)
ypred=knn.predict(xtest)
```

```python
print(classification_report(ytest,ypred))
```

```python
accuracy_score(ytest,ypred)
```

```python
confusion_matrix(ytest,ypred)
```

#### Hyper-Parameter Tuning

```python
accuracy=[]
for i in range(1,21):
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(xtrain,ytrain)
    ypred=knn.predict(xtest)
    ac=accuracy_score(ytest,ypred)
    accuracy.append(ac)
```

```python
plt.plot(range(1,21),accuracy)
plt.grid(True)
plt.show()
```

```python
knn=KNeighborsClassifier(n_neighbors=8)
knn.fit(xtrain,ytrain)
ypred=knn.predict(xtest)
```

# Deep Learning- Artificial Neural Network

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
ann=Sequential()

ann.add(Dense(units=9,activation="relu"))
ann.add(Dense(units=5,activation="relu"))
ann.add(Dense(units=1,activation="sigmoid"))
ann.compile(optimizer="adam",loss="binary_crossentropy",metrics="accuracy")
ann.fit(xtrain,ytrain,batch_size=100,validation_data=(xtest,ytest),epochs=800)
```

```python
lossdf=pd.DataFrame(ann.history.history)
lossdf.plot()
plt.show()
```

```python
#Low Bias 
#High Variance
#This is a case of Overfitting Model
```

```python
ann=Sequential()

ann.add(Dense(units=9,activation="relu"))
ann.add(Dense(units=5,activation="relu"))
ann.add(Dense(units=1,activation="sigmoid"))
ann.compile(optimizer="adam",loss="binary_crossentropy",metrics="accuracy")
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
earlystop=EarlyStopping( monitor="val_loss", mode="min", verbose=1, patience=80)
```

```python
ann.fit(xtrain,ytrain,batch_size=100,validation_data=(xtest,ytest),epochs=800, callbacks=[earlystop])
```

```python
lossdf=pd.DataFrame(ann.history.history)
lossdf.plot()
plt.show()
```

```python
from tensorflow.keras.layers import Dropout
```

```python
ann=Sequential()

ann.add(Dense(units=9,activation="relu"))
ann.add(Dropout(rate=0.2))
ann.add(Dense(units=5,activation="relu"))
ann.add(Dropout(rate=0.2))
ann.add(Dense(units=1,activation="sigmoid"))
ann.compile(optimizer="adam",loss="binary_crossentropy",metrics="accuracy")
```

```python
ann.fit(xtrain,ytrain,batch_size=100,validation_data=(xtest,ytest),epochs=800, callbacks=[earlystop])
```

```python
lossdf=pd.DataFrame(ann.history.history)
lossdf.plot()
plt.show()
```

```python
ypred=ann.predict(xtest)
```
